```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import re
from selenium.webdriver.chrome.options import Options
import os
import requests
```

```python
driver=webdriver.Chrome(executable_path="C:\Driver\chromedriver_win32 (3)\chromedriver.exe")
driver.get("https://www.igbc.ie/epd-search/")
driver.maximize_window()
```

```python
Dic = {}
# from selenium import webdriver
gc2= pd.DataFrame(data=Dic,index=[0],columns=['product_name','manufacturer','web','manufacturer_mail','manufacturer_website','manufacturer_address','epd_type','registration_date','Validity_untill','Publisher'])
```

```python
df=pd.read_excel(r'v.xlsx')
UR=df.web
for i in range(len(UR)):
    link=UR[i]
    driver.get(link)
    time.sleep(4)
    try:
        product_name=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > h4:nth-child(1)").text
    except:
        product_name="N\A"
    try:
        manufacturer=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(6) > div:nth-child(1) > div > div.col-sm-10").text
    except:
        manufacturer="N\A"
    try:
        manufacturer_mail=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(6) > div:nth-child(3) > div > div.col-sm-10 > a").text
    except:
        manufacturer_mail="N\A"
    try:
        manufacturer_website=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(6) > div:nth-child(4) > div > div.col-sm-10 > a").text
    except:
        manufacturer_website="N?A"
    try:
        manufacturer_address=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(6) > div:nth-child(5) > div > div.col-sm-10").text
    except:
        manufacturer_address="N/A"
    try:
        epd_type=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(4) > div > div:nth-child(1) > div:nth-child(1)").text.split("\n")[1]
    except:
        epd_type="N/A"
    ele=driver.find_elements_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(4) > div > div:nth-child(1) > div:nth-child(n)")
    len(ele)
    if len(ele)==4:
        try:
            registration_date=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(4) > div > div:nth-child(1) > div:nth-child(2)").text.split("\n")[1]
        except:
            registration_date="N/A"
        try:
            valid_until=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(4) > div > div:nth-child(1) > div:nth-child(3)").text.split("\n")[1]
        except:
            valid_until="N/A"
        try:
            epd_publisher=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(4) > div > div:nth-child(1) > div:nth-child(4)").text.split("\n")[1]
        except:
            epd_publisher="N/A"
    else:
        try:
            epd_publisher=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(4) > div > div:nth-child(1) > div:nth-child(2)").text.split("\n")[1]
        except:
            epd_publisher="N/A"
            
    Dic={'product_name':product_name,'manufacturer':manufacturer,'web':link,'manufacturer_mail':manufacturer_mail,'manufacturer_website':manufacturer_website,'manufacturer_address':manufacturer_address,'epd_type':epd_type,'registration_date':registration_date,'Validity_untill':valid_until,'Publisher':epd_publisher}
    gc2 = gc2.append(Dic, ignore_index=True)
    print(gc2)
    time.sleep(1)
    try:
        pdf=driver.find_element_by_css_selector("#main > div > div:nth-child(4) > div > div > div:nth-child(4) > div > div.col-sm-3.epd_single_image > a")

        URL=pdf.get_attribute("href")
        time.sleep(3)
        output_dir='.\epd'
        response=requests.get(URL)
        if response.status_code==200:
            file_path=os.path.join(output_dir,os.path.basename(URL))
            with open(file_path,'wb') as f:
                f.write(response.content)
    except:
        continue
    time.sleep(2)

```
