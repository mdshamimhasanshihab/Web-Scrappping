```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.support.ui import Select
```

```python
driver=webdriver.Chrome(executable_path="C:/Driver/cc/chromedriver.exe")
driver.get("https://www.shopwearwell.com/collections/all")
driver.maximize_window()
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0],columns=["Product_name","URL","price"])
```

```python
for j in range(1):
    product_name=driver.find_elements(By.CSS_SELECTOR,"#shopify-section-template--15307179589768__main > div > div > div > div > div > div.js-grid > div:nth-child(n) > div > div.product__grid__info.product__grid__info--under.text-center > a > div.product__grid__title__wrapper")
    product_link=driver.find_elements(By.CSS_SELECTOR,"#shopify-section-template--15307179589768__main > div > div > div > div > div > div.js-grid > div:nth-child(n) > div > div.product__grid__info.product__grid__info--under.text-center > a")
    price=driver.find_elements(By.CSS_SELECTOR,"#shopify-section-template--15307179589768__main > div > div > div > div > div > div.js-grid > div:nth-child(n) > div > div.product__grid__info.product__grid__info--under.text-center > a > div.product__grid__price")
    for i in range(len(product_name)):
        product=product_name[i].text
        link=product_link[i].get_attribute("href")
        pri=price[i].text
        Dic={"Product_name":product,"URL":link,"price":pri}
        gc = gc.append(Dic, ignore_index=True)
    print(gc)
    clk=driver.find_elements(By.CSS_SELECTOR,"#shopify-section-template--15307179589768__main > div > div > div > div > div > div.text-center > ul > li:nth-child(n) > a")
    clk=clk[-1]
    driver.execute_script("arguments[0].scrollIntoView();", clk)
    time.sleep(2)
    clk.click()
    time.sleep(2)
```

```python
df=pd.read_excel("Wearwell.xlsx")
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0],columns=["Product_name","URL","price","interest-free installments with shop pay",'save','size','Color','Description',"Details",'Care_Instruction'])
```

```python
for z in range(1,len(df)):
    driver.get(df.URL[z])
    time.sleep(3)
    product_url=df.URL[z]
    price=df.price[z]
    pr=df.Product_name[z]
    siz=""
    col=""
    try:
        save=driver.find_element(By.CLASS_NAME,"product__price--off").text.split(" ")[-1]
    except:
        save="NA"
    try:
        free_in=driver.find_element(By.CLASS_NAME,"shop-pay-terms").text.split("of")[-1].split("with")[0]
    except:
        free_in='NA'
    try:
        s=driver.find_elements(By.CLASS_NAME,"radio__fieldset")
        if len(s)>1:
            s=driver.find_elements(By.CLASS_NAME,"radio__fieldset")
            size=s[0].find_elements(By.CLASS_NAME,"radio__button")
            for i in size:
                siz=siz+", "+i.text
    except:
        siz='nA'
    try:
        color=driver.find_elements(By.CLASS_NAME,"radio__fieldset")[-1]
        colo=color.find_elements(By.CLASS_NAME,"radio__button")
        for j in colo:
            col=col+", "+j.text
    except:
        col="NA"
    try:
        clk=driver.find_element(By.CLASS_NAME,"accordion__title")
        driver.execute_script("arguments[0].scrollIntoView();", clk)
        time.sleep(1)
        clk.click()
        time.sleep(1)
        des=driver.find_element(By.CLASS_NAME,"accordion-content__entry").text
    except:
        continue
    try:
        clk=driver.find_elements(By.CLASS_NAME,"accordion__title")[1]
        driver.execute_script("arguments[0].scrollIntoView();", clk)
        time.sleep(1)
        clk.click()
        time.sleep(1)
        det=driver.find_element(By.CLASS_NAME,"metafield-multi_line_text_field").text
    except:
        continue
    try:
        cic=driver.find_elements(By.CLASS_NAME,"accordion__title")[2]
        driver.execute_script("arguments[0].scrollIntoView();", cic)
        time.sleep(1)
        cic.click()
        time.sleep(1)
        ci=driver.find_elements(By.CLASS_NAME,"metafield-multi_line_text_field")[1].text
    except:
        continue
    Dic = {"Product_name":pr,"URL":product_url,"price":price,"interest-free installments with shop pay":free_in,'save':save,'size':siz,'Color':col,'Description':des,"Details":det,'Care_Instruction':ci}
    gc = gc.append(Dic, ignore_index=True)
    print(gc)
    time.sleep(1)
```
