```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import re
from selenium.webdriver.chrome.options import Options
import os
import requests
```

```python
driver=webdriver.Chrome(executable_path="C:/Driver/cc/chromedriver.exe")
driver.get("https://companycheck.co.uk/search/?type=company-uk%2Ccompany-nltd&page=1")
```

```python
Dic={}
gc2= pd.DataFrame(data=Dic,index=[0],columns=['Company_Name','Address',"Result_Status",'Result_Number','Cash','Net_Worth','Asset','Liability','web_link'])
```

```python
for z in range(10):
    elemen=driver.find_elements_by_css_selector("#js-company-results > div:nth-child(n) > div.info-right")
    for i in range(len(elemen)):
        try:
            company_name=elemen[i].find_element_by_class_name("result__title").text
        except:
            company_name="N\A"
        try:
            web=elemen[i].find_element_by_class_name("result__title").get_attribute("href")
        except:
            web="N\A"
        try:
            address=elemen[i].find_element_by_class_name("result__address").text
        except:
            address="N\A"
        try:
            result_status=elemen[i].find_element_by_class_name("result__address").text
        except:
            result_status="N\A"
        try:
            result_number=elemen[i].find_element_by_class_name("result__number").text
        except:
            result_number="N\A"
        try:
            cash=elemen[i].find_element_by_class_name("cash").text.split(":")[1]
        except:
            cash="N\A"
        try:
            net_worth=elemen[i].find_element_by_class_name("net-worth").text.split(":")[1]
        except:
            net_worth="N\A"
        try:
            asset=elemen[i].find_element_by_class_name("assets").text.split(":")[1]
        except:
            asset="N\A"
        try:
            liability=elemen[i].find_element_by_class_name("liabilities").text.split(":")[1]
        except:
            liability="N\A"
        Dic={'Company_Name':company_name,'Address':address,"Result_Status":result_status,'Result_Number':result_number,'Cash':cash,'Net_Worth':net_worth,'Asset':asset,'Liability':liability,'web_link':web}
        gc2 = gc2.append(Dic, ignore_index=True)
        
    clk=driver.find_element_by_css_selector("#js-company-pagination > div.arrows > div.next.page-box")
    driver.execute_script("arguments[0].click();", clk)
    time.sleep(3)
    print(gc2)
```
