```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.support.ui import Select
```

```python
driver=webdriver.Chrome(executable_path="C:/Driver/cc/chromedriver.exe")
driver.get("https://projects.worldbank.org/en/projects-operations/projects-list?countrycode_exact=3A&os=0")
driver.maximize_window()
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0],columns=['Project_Title','Country','Project_id','Commitment_Amount',"Status",'Approval_date','Last_Updated','Last_Stage_Reached'])
```

```python
table=driver.find_element(By.CSS_SELECTOR,"body > div.par.parsys > div.project_site_spa.section > projectsall > div.par.parsys > div > projects-list > div > div > div > div > div > div.col-lg-9.col-md-9.col-sm-12.col-xs-12 > div:nth-child(4) > div > div.c14v1-body.c14v1-body-text.responsive-table.project-opt-table > div > table > tbody")
```

```python
tr=table.find_elements(By.CSS_SELECTOR,"tr")
```

```python
print(gc)
```

```python
for d in range(3,1084):
    table=driver.find_element(By.CSS_SELECTOR,"body > div.par.parsys > div.project_site_spa.section > projectsall > div.par.parsys > div > projects-list > div > div > div > div > div > div.col-lg-9.col-md-9.col-sm-12.col-xs-12 > div:nth-child(4) > div > div.c14v1-body.c14v1-body-text.responsive-table.project-opt-table > div > table > tbody")
    tr=table.find_elements(By.TAG_NAME,"tr")
    for t in range(0,len(tr)):
        try:
            project_title=tr[t].find_elements(By.TAG_NAME,"td")[0].text
        except:
            project_title=""
        try:
            Country=tr[t].find_elements(By.TAG_NAME,"td")[1].text
        except:
            Country=""
        try:
            project_id=tr[t].find_elements(By.TAG_NAME,"td")[2].text
        except:
            project_id=""
        try:
            Commitment_amount=tr[t].find_elements(By.TAG_NAME,"td")[3].text
        except:
            Commitment_amount=""
        try:
            Status=tr[t].find_elements(By.TAG_NAME,"td")[4].text
        except:
            Status=""
        try:
            Approval_date=tr[t].find_elements(By.TAG_NAME,"td")[5].text
        except:
            Approval_date=""
        try:
            Last_updated=tr[t].find_elements(By.TAG_NAME,"td")[6].text
        except:
            Last_updated=''
        try:
            last_stage_reached=tr[t].find_elements(By.TAG_NAME,"td")[7].text
        except:
            last_stage_reached=""
            
        Dic={'Project_Title':project_title,'Country':Country,'Project_id':project_id,'Commitment_Amount':Commitment_amount,"Status":Status,'Approval_date':Approval_date,'Last_Updated':Last_updated,'Last_Stage_Reached':last_stage_reached}
        gc = gc.append(Dic, ignore_index=True)
    try:    
        clk=driver.find_element(By.CSS_SELECTOR,"body > div.par.parsys > div.project_site_spa.section > projectsall > div.par.parsys > div > projects-list > div > div > div > div > div > div.col-lg-9.col-md-9.col-sm-12.col-xs-12 > div:nth-child(5) > div > ul > li:nth-child(13)")
        clk.click()
        time.sleep(3)
    except:
        driver.refresh()
        time.sleep(3)
        clk=driver.find_element(By.CSS_SELECTOR,"body > div.par.parsys > div.project_site_spa.section > projectsall > div.par.parsys > div > projects-list > div > div > div > div > div > div.col-lg-9.col-md-9.col-sm-12.col-xs-12 > div:nth-child(5) > div > ul > li:nth-child(13)")
        clk.click()
        time.sleep(4)
        
    print(gc)
```
