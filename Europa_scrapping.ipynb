```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import re
```

```python
driver=webdriver.Chrome(executable_path="C:/Driver/cc/chromedriver.exe")
driver.get("https://ec.europa.eu/eurostat/ramon/relations/index.cfm?TargetUrl=LST_LINK&StrNomRelCode=PRODCOM%202008%20-%20CN%202008&StrLanguageCode=EN")
```

```python
Dic = {}
# from selenium import webdriver
gc1= pd.DataFrame(data=Dic,index=[0],columns=['Source_code','Source_Product','Target_code','Target_product'])
```

```python

for page in range(71,220):
    tablerow=driver.find_elements_by_css_selector("#wrapper > form:nth-child(4) > table > tbody > tr:nth-child(8) > td > table > tbody > tr:nth-child(n)")
    for i in range(1,len(tablerow)):
        target_text=tablerow[i].find_elements_by_class_name("text")
        source=target_text[0].text.split("\n")
        source_code=source[0]
        source_product=source[1]
        source1=target_text[2].text.split("\n")
        target_code=source1[0]
        target_product=source1[1]
        Dic={'Source_code':source_code,'Source_Product':source_product,'Target_code':target_code,'Target_product':target_product}
        gc1 = gc1.append(Dic, ignore_index=True)
    
    print(gc1)
#     time.sleep(1)

    clk=driver.find_element(By.CSS_SELECTOR,"#wrapper > form:nth-child(4) > table > tbody > tr:nth-child(10) > td:nth-child(2) > input:nth-child(4)")
    driver.execute_script("arguments[0].click();", clk)
#     time.sleep(2)
```
