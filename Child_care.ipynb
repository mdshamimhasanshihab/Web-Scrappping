```python
for b in range(23):
    tbody=driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_rgSearchResults_ctl00 > tbody")
    td=tbody.find_elements(By.TAG_NAME,"a")
    print(len(td))
    for a in range(len(td)):
        tbody=driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_rgSearchResults_ctl00 > tbody")
        td=tbody.find_elements(By.TAG_NAME,"a")
        if a%2==0:
            print(a)
            clk=td[a]
            driver.execute_script("arguments[0].click();",clk)
            time.sleep(3)
            co_name=driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_FacilityDetail_rptBasicFacilityInfo_FacilityNameLabel_0").text
            address_1=driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_FacilityDetail_rptBasicFacilityInfo_FacilityStreetLabel_0").text
            city=driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_FacilityDetail_rptBasicFacilityInfo_FacilityCityLabel_0").text
            state=driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_FacilityDetail_rptBasicFacilityInfo_FacilityStateLabel_0").text
            zipp=driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_FacilityDetail_rptBasicFacilityInfo_FacilityZipLabel_0").text
            email=driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_FacilityDetail_Accordion1 > div:nth-child(3) > div > div > div:nth-child(4) > div:nth-child(2)").text
            try:
                date=driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_FacilityDetail_rptBasicFacilityInfo_InspectionDateLabel_0").text
            except:
                date=""
            driver.execute_script("window.scrollTo(0, 800)")
            time.sleep(2)
            driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_FacilityDetail_Accordion1 > div:nth-child(8)").click()
            time.sleep(2)
            
            owner=driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_FacilityDetail_lblOwnerName").text
            f_name=owner.split(" ")[0]
            l_name=owner.split(" ")[-1]
            driver.execute_script("window.scrollTo(0, 0)")
            time.sleep(1)
            driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_btnReturnToList").click()
            time.sleep(3)
            Dic={'Company':co_name,'Address1':address_1,'City':city,'State':state,'Zip code':zipp,'Email':email,'Contact_Person':owner,'First_Name':f_name,'Last_Name':l_name,'Date':date}
            gc = gc.append(Dic, ignore_index=True)
            print(gc)
    driver.execute_script("window.scrollTo(0, 0)")
    driver.find_element(By.CSS_SELECTOR,"#dnn_ctr1464_View_rgSearchResults_ctl00 > thead > tr.rgPager > td > div > div.rgWrap.rgArrPart2 > button.t-button.rgActionButton.rgPageNext > span").click()
    time.sleep(2)
        
```
