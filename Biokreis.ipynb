```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.support.ui import Select
```

```python
driver=webdriver.Chrome(executable_path="C:/Driver/cc/chromedriver.exe")
driver.get("https://www.biokreis.de/bio-fuer-dich/bio-in-deiner-region/")
driver.maximize_window()
```

```python
company=driver.find_elements(By.CSS_SELECTOR,"#wpgmza_table_15 > tbody > tr:nth-child(n) > td.wpgmza_table_title > font > font")
product=driver.find_elements(By.CSS_SELECTOR,"#wpgmza_table_15 > tbody > tr:nth-child(n) > td.wpgmza_table_title > span")
# for i in range(len(company)):
#     com=
len(company)
```

```python
Dic= {}
gc= pd.DataFrame(data=Dic,index=[0],columns=["Company","Address","Opening hour","Contact"])
```

```python
table=driver.find_element(By.CSS_SELECTOR,"#wpgmza_table_15 > tbody")
tr=table.find_elements(By.TAG_NAME,"tr")
for i in range(1,len(tr)):
    company=tr[i].find_elements(By.TAG_NAME,"td")[1].text
    address=tr[i].find_elements(By.TAG_NAME,"td")[3].text
    oh=tr[i].find_elements(By.TAG_NAME,"td")[7].text
    contact=tr[i].find_elements(By.TAG_NAME,"td")[-2].text
    Dic={"Company":company,"Address":address,"Opening hour":oh,"Contact":contact}
    gc= gc.append(Dic, ignore_index=True)
    print(gc)
```
