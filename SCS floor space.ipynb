```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import re
```

```python
driver=webdriver.Chrome(executable_path="C:\Driver\chromedriver_win32 (3)\chromedriver.exe")
driver.get("https://www.scsglobalservices.com/services/floorscore")
driver.maximize_window()
```

```python
Dic2 = {}
Dic1={}
# from selenium import webdriver
gc2= pd.DataFrame(data=Dic2,index=[0],columns=['Company','Type','Category','web'])
gc1= pd.DataFrame(data=Dic1,index=[0],columns=['web'])
```

```python
listt=driver.find_elements_by_css_selector("#block-view-services-menu > div > div > ul > li > ul > li:nth-child(n)")
for i in listt:
    service=i.find_element_by_tag_name("a").get_attribute("href")
    Dic1={"web":service}
    gc1 = gc1.append(Dic1, ignore_index=True)
```

```python
print(gc1)
```

```python
driver.find_element_by_link_text("Services").click()
```

```python
for i in range(len(gc1.web)):
    time.sleep(2)
    service=gc1.web[i+1]
    driver.get(service)
    time.sleep(2)
    category=driver.find_element_by_tag_name("h1").text
    xx=driver.find_elements_by_css_selector("#block-scsglobalservices-system-main > div > div > div.category-container.clearfix > div.right-col > div.taxonomy-content > div.item-list.clearfix > div > ul > li:nth-child(n)")
    for j in xx:
        try:
            c_name=(j.text.split("\n")[0])
            c_type=(j.text.split("\n")[1])
            web=j.find_element_by_class_name("service-hover").find_element_by_tag_name("a").get_attribute("href")

            Dic2={'Company':c_name,'Type':c_type,'Category':category,'web':web}
            gc2 = gc2.append(Dic2, ignore_index=True)
            print(gc2)
        except:
            continue
```
