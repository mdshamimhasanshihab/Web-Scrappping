```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.support.ui import Select
```

```python
driver=webdriver.Chrome(executable_path="C:/Driver/cc/chromedriver.exe")
driver.get("https://www.fertigbau.de/mitglieder/fertighaushersteller/a.html")
driver.maximize_window()
```

```python
cr=driver.find_elements(By.CLASS_NAME,"teaser_more")
cr1=cr[1].find_element(By.TAG_NAME,"a").get_attribute("href")
```

```python
driver.get(cr1)
phone=driver.find_element(By.CLASS_NAME,"telefon").text
fax=driver.find_element(By.CLASS_NAME,"fax").text
website=driver.find_element(By.CLASS_NAME,"internet").text
mail=driver.find_element(By.CLASS_NAME,"mail").text
print(phone)
print(fax)
print(website)
print(mail)
```

```python
Dic1 = {}
gc1= pd.DataFrame(data=Dic1,index=[0],columns=["link"])
```

```python
linkkk=driver.find_element(By.CLASS_NAME,"mitgliederaz")
linkk=linkkk.find_elements(By.TAG_NAME,'span')
# print(linkk)
for li in range(len(linkk)):
    try:
        link=linkk[li].find_element(By.TAG_NAME,'a').get_attribute("href")
        Dic1={'link':link}
        gc1 = gc1.append(Dic1, ignore_index=True)
    except:
        continue

```

```python
print(gc1)
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0],columns=["member","addresss",'phone','fax','website','mail'])
```

```python
for z in range(1,len(gc1)):
    driver.get(gc1.link[z])
    cr=driver.find_elements(By.CLASS_NAME,"teaser_more")
    for i in range(len(cr)):
        
        name=driver.find_elements(By.TAG_NAME,"h2")
        address=driver.find_elements(By.TAG_NAME,"b")
        cr=driver.find_elements(By.CLASS_NAME,"teaser_more")
        member=name[i].text
        adds=name[i].text
        try:
            cr1=cr[i].find_element(By.TAG_NAME,"a").get_attribute("href")
        except:
            continue
        driver.get(cr1)
        try:
            phone=driver.find_element(By.CLASS_NAME,"telefon").text
        except:
            phone="NA"
        try:
            fax=driver.find_element(By.CLASS_NAME,"fax").text
        except:
            fax="NA"
        try:
            website=driver.find_element(By.CLASS_NAME,"internet").text
        except:
            website="NA"
        try:
            mail=driver.find_element(By.CLASS_NAME,"mail").text
        except:
            mail="NA"
        driver.back()
        time.sleep(2)
        Dic={"member":member,"addresss":adds,'phone':phone,'fax':fax,'website':website,'mail':mail}
        gc = gc.append(Dic, ignore_index=True)
        print(gc)
```
