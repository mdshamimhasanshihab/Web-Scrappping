```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.support.ui import Select
```

```python
driver=webdriver.Chrome(executable_path="C:/Driver/cc/chromedriver.exe")
driver.get("https://www.tsacg.com/individual/plan-sponsor/")
driver.maximize_window()
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0],columns=['Name','Security_Benefit',"GWN",'Orion','Aspire','Fidelity','State'])
```

```python
trr=driver.find_elements(By.CSS_SELECTOR,"#form > div.main.cf > div.col-content.full-width > article > div.list.border-btm > ul > li:nth-child(n) > a")
for z in range(len(trr)):
    trr=driver.find_elements(By.CSS_SELECTOR,"#form > div.main.cf > div.col-content.full-width > article > div.list.border-btm > ul > li:nth-child(n) > a")
    state=trr[z].text
    trr[z].click()
    time.sleep(2)
    tr=driver.find_elements(By.CSS_SELECTOR,"#form > div.main.cf > div.col-content.full-width > article > div.list.border-btm > ul > li:nth-child(n) > a")
    for k in range(len(tr)):
        aspire=""
        fidelity=""
        gwn=""
        orion=""
        security=""    
        tr=driver.find_elements(By.CSS_SELECTOR,"#form > div.main.cf > div.col-content.full-width > article > div.list.border-btm > ul > li:nth-child(n) > a")
        name=tr[k].text
        tr[k].click()
        time.sleep(2)
        td=driver.find_elements(By.CSS_SELECTOR,"#form > div.main.cf > div.col-content > article > ul > ul > li > div > ul > li:nth-child(1) > ul > li:nth-child(n) > a")
        for i in range(len(td)):
            emp=td[i].text
            emp_part=emp.split(" ")[0]
            if emp_part=="Aspire":
                aspire="ok"
            if emp_part=="ASPire":
                aspire="ok"
            if emp_part=="Fidelity":
                fidelity="ok"
            if emp_part=="GWN":
                gwn="ok"
            if emp_part=="Orion":
                orion="ok"
            if emp_part=="Security":
                security="ok"
        Dic={'Name':name,'Security_Benefit':security,"GWN":gwn,'Orion':orion,'Aspire':aspire,'Fidelity':fidelity,'State':state}
        gc = gc.append(Dic, ignore_index=True)       
        print(gc)
        driver.back()
        time.sleep(1)
    driver.get("https://www.tsacg.com/individual/plan-sponsor/")
    time.sleep(10)
```

```python
gc.to_excel("New_Jercy_solution.xlsx")
```
