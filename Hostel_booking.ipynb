```python
import requests
import json
import time


# User credentials
email = 'gorfein1@gmail.com'
password = '!Test2023!'
restaurant_id = '68493'  # Example restaurant ID
party_size = 2
date = '2023-04-04'  # Format: YYYY-MM-DD
time1 = '19:00:00'  # Format: HH:mm:ss


# Step 1: Obtain an access token
def get_access_token(email, password):
    url = 'https://api.resy.com/3/auth/password'
    headers = {'Content-Type': 'application/json'}
    data = json.dumps({'email': email, 'password': password})
    response = requests.post(url, headers=headers, data=data)

    #return response
    if response.status_code == 200:
        return response.json()['access_token']
    else:
        raise Exception(f'Error obtaining access token: {response.text}')

# Step 2: Get available slots
def get_availability(restaurant_id, party_size, date, access_token):
    url = f'https://api.resy.com/4/find?restaurant_id={restaurant_id}&party_size={party_size}&date={date}'
    headers = {'Authorization': f'Bearer {access_token}'}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()['results']['availability']['slots']
    else:
        raise Exception(f'Error getting availability: {response.text}')

# Step 3: Book a reservation
def book_reservation(slot_id, party_size, access_token):
    url = 'https://api.resy.com/4/reservation'
    headers = {
        'Content-Type': 'application/json',
        'Authorization': f'Bearer {access_token}'
    }
    data = json.dumps({'slot_id': slot_id, 'party_size': party_size})
    response = requests.post(url, headers=headers, data=data)
    if response.status_code == 200:
        return response.json()['reservation']
    else:
        raise Exception(f'Error booking reservation: {response.text}')


if __name__ == '__main__':

    secured_reservation = False

    # Main code
    while not secured_reservation:

        try:
            access_token = get_access_token(email, password)
            slots = get_availability(restaurant_id, party_size, date, access_token)
            if slots:
                slot_id = slots[0]['id']  # Use the first available slot
                reservation = book_reservation(slot_id, party_size, access_token)
                print(f'Reservation booked successfully! Details: {reservation}')
                secured_reservation = True
            else:
                print('No available slots found for the given date and time.')
        except Exception as e:
            print(f'Error: {e}')


```
