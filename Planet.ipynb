```python
pdf=df.Website
```

```python
import os
import requests
for i in range(len(pdf)):
    hrf=pdf[i+199]
    URL=hrf
    output_dir='.\Outputs'
    try:
        response=requests.get(URL)
    except:
        continue
    if response.status_code==200:
        file_path=os.path.join(output_dir,os.path.basename(URL))

        with open(file_path,'wb') as f:
            f.write(response.content)
            time.sleep(5)
```
