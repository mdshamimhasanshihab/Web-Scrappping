```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.support.ui import Select
```

```python
driver=webdriver.Chrome(executable_path="C:/Driver/cc/chromedriver.exe")
driver.get("https://www.iafcertsearch.org/search/certification-bodies")
driver.maximize_window()
```

```python
certification=driver.find_elements(By.CSS_SELECTOR,"#root > div > div > div > div > div > div > div > div.my-5.container > div.d-flex.row > div:nth-child(n)")
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0],columns=["certification_name",'country','description','url'])
```

```python
for a in range(63):
    certification=driver.find_elements(By.CSS_SELECTOR,"#root > div > div > div > div > div > div > div > div.my-5.container > div.d-flex.row > div:nth-child(n)")
    for i in range(len(certification)):
        name=certification[i].find_element(By.TAG_NAME,"h5").text
        country=certification[i].find_element(By.TAG_NAME,"h6").text
        des=certification[i].find_element(By.TAG_NAME,"p").text
        url=certification[i].find_element(By.TAG_NAME,"a").get_attribute("href")
        Dic={"certification_name":name,'country':country,'description':des,'url':url}
        gc = gc.append(Dic, ignore_index=True)
    print(gc)
    driver.find_elements(By.CLASS_NAME,"page-item")[-1].click()
    time.sleep(2)
    
```

```python
gc.to_excel("IAF.xlsx")
```

```python
df=pd.read_excel("IAF.xlsx")
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0],columns=["certification_name",'country','description','accredation_company','accredation_company_link','accredation_standard','Tecnical_sector','Country/economy','Office_directories'])
```

```python
for c in range(1232,1240):
    driver.get(url[c])
    time.sleep(4)
    cert=df.certification_name[c]
    country=df.country[c]
    des=df.description[c]
    acc=driver.find_element(By.TAG_NAME,"h5")
    accredation_company=acc.text
    accredated_com_link=acc.find_element(By.TAG_NAME,"a").get_attribute("href")
    accredation_standard=driver.find_elements(By.CLASS_NAME,"card-body")[0].text
    ts=driver.find_elements(By.CLASS_NAME,"card-body")[1].text
    ce=driver.find_elements(By.CLASS_NAME,"card-body")[2].text
    od=driver.find_elements(By.CLASS_NAME,"card-body")[3].text
    Dic={"certification_name":cert,'country':country,'description':des,'accredation_company':accredation_company,'accredation_company_link':accredated_com_link,'accredation_standard':accredation_standard,'Tecnical_sector':ts,'Country/economy':ce,'Office_directories':od}
    gc = gc.append(Dic, ignore_index=True)
    print(gc)
```
