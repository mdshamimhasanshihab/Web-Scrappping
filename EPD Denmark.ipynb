```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import re
from selenium.webdriver.chrome.options import Options
import os
import requests
```

```python
Dic = {}
# from selenium import webdriver
gc2= pd.DataFrame(data=Dic,index=[0],columns=['product_name','details','web','pdf_link','issue_date','validity','Declared/Functional Unit','Number of declared datasets in the EPD'])
```

```python
df=pd.read_excel(r'C:/Users/User/OneDrive/Desktop/Data_acq/u.xlsx')
```

```python
link=df.Web
```

```python
driver=webdriver.Chrome(executable_path="C:\Driver\chromedriver_win32 (3)\chromedriver.exe")
for i in range(len(link)):
    driver.get(link[i+160])
    pdf=driver.find_element_by_css_selector("body > section > div:nth-child(2) > div > div > div:nth-child(1) > div.small-12.medium-9.large-9.columns > a")
    URL=pdf.get_attribute("href")
    time.sleep(3)
    output_dir='.\Outputs'
    response=requests.get(URL)
    if response.status_code==200:
        file_path=os.path.join(output_dir,os.path.basename(URL))
        with open(file_path,'wb') as f:
            f.write(response.content)
    time.sleep(6)
    try:
        issue_date=driver.find_element_by_css_selector("body > section > div:nth-child(2) > div > div > div:nth-child(2) > div.small-12.medium-9.large-9.columns").text
    except:
        issue_date="N/A"
    try:
        validity=driver.find_element_by_css_selector("body > section > div:nth-child(2) > div > div > div:nth-child(3) > div.small-12.medium-9.large-9.columns").text
    except:
        validity="N/A"
    try:
        dfu=driver.find_element_by_css_selector("body > section > div:nth-child(2) > div > div > div:nth-child(4) > div.small-12.medium-9.large-9.columns").text
    except:
        dfu="N/A"
    try:
        ndd=driver.find_element_by_css_selector("body > section > div:nth-child(2) > div > div > div:nth-child(5) > div.small-12.medium-9.large-9.columns").text
    except:
        ndd="N/A"
    Dic={'product_name':df.Product_name[i+160],'details':df.Details[i+160],'web':link[i+160],'pdf_link':URL,'issue_date':issue_date,'validity':validity,'Declared/Functional Unit':dfu,'Number of declared datasets in the EPD':ndd}
    gc2 = gc2.append(Dic, ignore_index=True)
    print(gc2)
```

```python
gc2.to_excel(r'C:/Users/User/OneDrive/Desktop/Data_acq/Exported_File/Product_File/EPD Denmark_13_07_22.xlsx')
```
