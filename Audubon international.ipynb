```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import re
```

```python
driver=webdriver.Chrome(executable_path="C:/Driver/cc/chromedriver.exe")
driver.get("https://directory.auduboninternational.org/directory")
```

```python
Dic = {}
# from selenium import webdriver
gc= pd.DataFrame(data=Dic,index=[0],columns=['Company','Address','Region','Certification_date','web'])
```

```python
for z in range(39):
    dictionary=driver.find_elements_by_class_name("directory-card")
    for i in range(len(dictionary)):
        company=dictionary[i].find_element_by_class_name("member-title").text
        Address=dictionary[i].find_element_by_tag_name("p").text
        Region=dictionary[i].find_element_by_css_selector("#fl-main-content > div.neon-custom-app-content > div > div.row.my-3.flex-row > div:nth-child(n) > div > p:nth-child(3)").text
        certification_date=dictionary[i].find_element_by_css_selector("#fl-main-content > div.neon-custom-app-content > div > div.row.my-3.flex-row > div:nth-child(n) > div > p:nth-child(4)").text.split(":")[1]
        try:
            web=dictionary[i].find_element_by_tag_name("a").get_attribute("href")
        except:
            web="N/A"
        Dic={'Company':company,'Address':Address,'Region':Region,'Certification_date':certification_date,'web':web}
        gc = gc.append(Dic, ignore_index=True)
        print(gc)
    nextt=driver.find_element_by_css_selector("#fl-main-content > div.neon-custom-app-content > div > div.neon-pagination.row.my-3 > div.col-sm-4.text-right > a").get_attribute("href")
    driver.get(nextt)
    time.sleep(3)
```

```python
gc.to_excel(r'Audubon international.xlsx')
```
