```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0],columns=['Company','Website','Country','Status',"Description",'Archives'])
```

```python
for b in range(1000):   
    for a in range(4):
        btn=driver.find_element(By.CLASS_NAME,"vc_tta-tabs-container").find_elements(By.TAG_NAME,"div")
        btn[a].click()
        try:
            company=btn[a].get_attribute("id").split("-")[-1]
        except:
            company="NA"
        try:
            link=driver.find_element(By.CLASS_NAME,"vc_tta-panels-container").text
        except:
            link="NA"
        try:
            website=link.split("\n")[0].split(" ")[1]
        except:
            website:"NA"
        try:
            country=link.split("\n")[1].split(":")[1]
        except:
            country=link.split("\n")[1]
        try:
            status=link.split("\n")[2].split(":")[1]
        except:
            status="NA"
        try:
            des=link.split("\n")[3]
        except:
            des="NA"
        try:
            arc=link.split("\n")[4].split(":")[1]
        except:
            arc="NA"
        Dic={'Company':company,'Website':website,'Country':country,'Status':status,"Description":des,'Archives':arc}
        gc = gc.append(Dic, ignore_index=True)
        print(gc)
        time.sleep(1)
    btn=driver.find_element(By.ID,"btn-next-miembros")
    btn.click()
    time.sleep(3)
```

```python
gc.to_excel("Responsible soy_15_09_22_Kazi.xlsx")
```
