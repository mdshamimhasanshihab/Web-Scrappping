```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.support.ui import Select
```

```python
driver=webdriver.Chrome(executable_path="C:/Driver/cc/chromedriver.exe")
driver.get("https://www.maaturism.ee/index.php?id=ruraltourism-search&cat_id=0&onpage=80")
driver.maximize_window()
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0],columns=["spot","addresss",'description','website'])
```

```python
for b in range(8,9):
    ele=driver.find_elements(By.CLASS_NAME,"filters-results-item")
    
    for a in range(len(el)):
        web=ele[a].find_element(By.CLASS_NAME,'content').find_element(By.TAG_NAME,'a').get_attribute("href")
        spot=ele[a].find_element(By.TAG_NAME,'h3').text
        address=ele[a].find_element(By.CLASS_NAME,'location').text
        des=ele[a].find_element(By.CLASS_NAME,'desc').text
        Dic={"spot":spot,"addresss":address,'description':des,'website':web}
        gc = gc.append(Dic, ignore_index=True)
        print(gc)
    driver.find_element(By.LINK_TEXT,f'{b}').click()
```

```python
df=pd.read_excel("Estonian eco turism quality label.xlsx")
url=df.website
```

```python
gc.to_excel("Estonian eco turism quality label_14_09_22_kazi.xlsx")
```

```python
Dic = {}
gc= pd.DataFrame(data=Dic,index=[0],columns=["spot","addresss",'description','info_link','contact_no','mail','website','location_on_map','facility'])
```

```python
for d in range(1,len(url)):
    driver.get(url[d])
    try:
        contact=driver.find_element(By.CSS_SELECTOR,"body > div.main-content-wrap > div > div > div > div > div.contact-card > div.contact-card-row.contact-card-row2.row > div:nth-child(2) > p").text
    except:
        contact="NA"
    try:
        e_contact=driver.find_element(By.CSS_SELECTOR,"body > div.main-content-wrap > div > div > div > div > div.contact-card > div.contact-card-row.contact-card-row2.row > div:nth-child(3) > p").text
    except:
        e_contact="NA"
    try:
        mail=e_contact.split("\n")[0]
    except:
        mail="NA"
    try:
        website=e_contact.split("\n")[1]
    except:
        website="NA"
    try:
        location_on_map=driver.find_element(By.CSS_SELECTOR,"body > div.main-content-wrap > div > div > div > div > div.contact-card > div.contact-card-row.contact-card-row2.row > div:nth-child(1) > p > a").get_attribute("href")
    except:
        location_on_map="NA"
    try:
        facility=driver.find_element(By.CSS_SELECTOR,"body > div.main-content-wrap > div > div > div > div > div.company-opts-filter > div").text
    except:
        facility="NA"
    Dic={"spot":df.spot[d],"addresss":df.addresss[d],'description':df.description[d],'info_link':url[d],'contact_no':contact,'mail':mail,'website':website,'location_on_map':location_on_map,'facility':facility}
    gc = gc.append(Dic, ignore_index=True)
    print(gc)
    time.sleep(1)
```
